```python
import ujson as json
import requests
import os

timeout = 60

TRAVEL_HOST = "128.111.28.87"
TRAVEL_PORT = 8830
# TRAVEL_HOST = os.getenv("TRAVEL_HOST", "localhost")
# TRAVEL_PORT = os.getenv("TRAVEL_PORT", 10300)
```

```python
def test(argument: str):
    url = f"http://{TRAVEL_HOST}:{TRAVEL_PORT}/{argument}"
    resp = requests.get(url, timeout=timeout)
    print(f"response:\n{json.dumps(resp.json(), indent=4)}")
```

```python
test("query_accommodation?city=San Francisco")
```

```python
test("query_attraction?city=San Diego")
```

```python
test("query_city?state=California")
```

```python
test("query_flight?origin=San Francisco&destination=Los Angeles&departure_date=2022-04-04")
```

```python
test("query_distance?origin=San Francisco&destination=Los Angeles")
```
